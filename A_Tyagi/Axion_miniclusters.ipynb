```python
# Define the QCD transition temperature
def calc_mu(m_a, f_a):
    return  np.sqrt(m_a/1e9 * f_a) #*2.5 

# Define the temperature-dependent mass ma(T) in GeV
def calc_ma(T, m_a, f_a, n):
    mu = calc_mu(m_a, f_a)
    return m_a/1e9 * (T / mu) ** (-n) if T > mu else m_a/1e9

# Define the Hubble parameter H(T)
def H(T, g_star_r):
    return np.sqrt(np.pi**2  * g_star_r * T**4 / (90* M_pl**2))
```

```python
# Define the function to calculate g_star
def g_star_R(T):
    # print ('calculating g_star_R')
    a0 = 1.21
    a_j1 = np.array([0.572, 0.330, 0.579, 0.138, 0.108])
    a_j2 = np.array([-8.77, -2.95, -1.80, -0.162, 3.76])
    a_j3 = np.array([0.682, 1.01, 0.165, 0.934, 0.869])

    # for T in Temp:
    t = np.log(T / 1.0)
    sum_terms = [a_j1[i] * (1.0 + np.tanh((t - a_j2[i]) / a_j3[i])) for i in range(5)]
    # print(f"dum terms: {sum_terms}")
    g_star_r_value = np.exp(a0 + np.sum(sum_terms))

    return g_star_r_value # np.array(result)

def g_star_S(T):
    # print ('calculating g_star_S')
    a0 = 1.36
    a_j1 = np.array([0.498, 0.327, 0.579, 0.140, 0.109])
    a_j2 = np.array([-8.74, -2.89, -1.79, -0.102, 3.82])
    a_j3 = np.array([0.693, 1.01, 0.155, 0.963, 0.907])

    # for T in Temp:
    t = np.log(T / 1.0)
    sum_terms = [a_j1[i] * (1.0 + np.tanh((t - a_j2[i]) / a_j3[i])) for i in range(5)]
    # print(f"dum terms: {sum_terms}")
    g_star_s_value = np.exp(a0 + np.sum(sum_terms))

    return g_star_s_value # np.array(result)
```

```python
# Temperature range
Temp = np.logspace(-6, 3, 500)  # T in GeV

# Calculate g_star_R and g_star_S
g_star_r = [g_star_R(T) for T in Temp]
g_star_s = [g_star_S(T) for T in Temp]

# Plotting
plt.figure(figsize=(7, 4))
plt.plot(Temp, g_star_r, label=r'$g_{\star R}$ fit', color='b')
plt.plot(Temp, g_star_s, label=r'$g_{\star S}$ fit', color='r')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Temperature (GeV)')
plt.ylabel(r'$g_{\star}$')
plt.title(r'$g_{\star R}$ and $g_{\star S}$ vs Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Solve 3H(T) = ma(T)
def find_Tosc(m_a, f_a, n):

    T_initial_guess = 10000 # Initial guess for T_osc in GeV
    
    # Function to find the root of
    def equation(T):
        g_star_r = g_star_R(T)
        # print(f"n = {n}, fa: {f_a}, g_star_R:{g_star_R}, T_osc {T[0]}")
        return 3 * H(T, g_star_r) - calc_ma(T, m_a, f_a, n)
    
    T_osc = fsolve(equation, T_initial_guess)[0]

    return T_osc
```

```python
# Plot T_osc vs f_a for different temperature coefficients n
mu = 0.2
def f_a(m_a_values):
    return (mu)**2 * 1e9 / m_a_values  # GeV
def m_a(f_a_values):
    return (mu)**2 * 1e9 / f_a_values 

m_a_values = np.logspace(-13, 0, 1000) #in eV
f_a_values = f_a(m_a_values)
f_a_limit = 8.2e12 #Gev from Inflation
m_a_limit = m_a(f_a_limit)


# Values of n to consider
n_values = [0, 3.34, 6]

# Plotting
fig, ax1 = plt.subplots(figsize=(8, 4))

for n in n_values:
    print(f'working on n = {n}')
    T_osc_values = [find_Tosc(m_a, f_a_values[i], n) for i, m_a in enumerate(m_a_values)]
    ax1.plot(m_a_values, T_osc_values, label=f'n={n}')
    ax2 = ax1.secondary_xaxis('top', functions=(f_a, m_a))

ax1.axvline(x=m_a_limit, color='black', linestyle='--', label=f'$f_a = {f_a_limit:.2e}$ GeV')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel(r'$m_a$ (eV)')
ax2.set_xlabel('$f_a$ (GeV)')
ax1.set_ylabel(r'$T_{osc}$ (GeV)')
plt.title(r'$T_{osc}$ vs $m_a$ for different n   ($T_c$ = $\sqrt{m_a f_a}$ = 0.2 GeV)', pad=15)

plt.legend()
plt.grid(True)
# plt.savefig(fr"output\T_osc vs m_a.png", dpi = 1000, bbox_inches='tight')
plt.show()
```

```python
# Define constants
M_pl = 1.22e18  # Planck mass in GeV

T_cmb = 2.725 # K
K_to_GeV = 8.6173e-14
T_CMB = T_cmb * K_to_GeV # CMB temperature in GeV

s_inv_to_ev = 6.58e-16
h = 0.7 #dimensionless Hubble constant parameter
H_0 = 100*h # km/s/Mpc
mpc_to_km = 3.086e19
H_0 = H_0/ mpc_to_km * s_inv_to_ev /1e9  # Hubble constant in GeV (H0 in km/s/Mpc converted to GeV)

omega_a_h2 =0.12
omega_a = omega_a_h2/h**2
print (H_0, T_CMB, omega_a)

alpha_dec = 2.48


# Define the function to calculate f_a for a given m_a
def calc_f_a(f_a, m_a, n, cn, alpha_d = alpha_dec):
    # mu = calc_mu(m_a, f_a)
    T_osc = find_Tosc(m_a, f_a, n)
    ma_Tosc = calc_ma(T_osc, m_a, f_a, n)
    ma_Tcmb = m_a/1e9 # calc_ma(T_CMB, m_a, f_a, n)
    a_Tosc = ((g_star_S(T_CMB)**(1/3)* T_CMB )/ (g_star_S(T_osc)**(1/3) *T_osc))
    result = (omega_a* 6 * H_0**2 * M_pl**2/ (1 + alpha_d)) - ( (np.pi**2 * cn / 3)* ma_Tcmb * ma_Tosc * f_a**2  * a_Tosc**3 )
    # print(f"ma,n,cn:{m_a, n, cn}, mu:{mu}, T_osc:{T_osc}, ma_Tosc:{ma_Tosc}, a_Tosc:{a_Tosc}, \nold fa:{f_a}, result = {result}") 
    return result
```

```python
M_pl = 2.3e18
# M_pl =1.22e19

# Define the range of m_a values (in eV)
m_a_values = np.logspace( -2, -15, 200)  # from 10^-11 to 10^-4 eV

# Define n and cn values
n_values = [0,  3.34, 6]
cn_values = [2.7, 2.1, 2.0]
f_a_limit = 8.2e12
factor = 1 #30

# Calculate f_a for each m_a
f_a_values = {}
for n, cn in zip(n_values, cn_values):
    f_a_values[n] = []
    print(f"Working for n: {n}")
    for m_a in m_a_values:
        fa_initial_guess = 1e11  # Initial guess for f_a in GeV
        fa_solution, = fsolve(calc_f_a, fa_initial_guess, args=(m_a, n, cn))
        # print(f"ma,n,cn:{m_a, n, cn}, fa_solution: {fa_solution}")
        if n == 0:
            if fa_solution/factor > f_a_limit:
                
                print(f"Breaking for n = {n}, m_a = {ma_last}, F_a = {fa_solution}")
                break   
            ma_last = m_a
            f_a_values[n].append(fa_solution/factor)
        else:
            if fa_solution/factor > f_a_limit:
                print(f"Breaking for n = {n}, m_a = {ma_last}, F_a = {fa_solution}")
                break
            ma_last = m_a
            f_a_values[n].append(fa_solution/factor)

    # print( m_a_values[len(f_a_values[n])-1])


f_a_values_approx ={}
for n in n_values:
    f_a_values_approx[n] = []
    print(f"Working for approx n: {n}")
    for m_a in m_a_values:
        fa_initial_guess = 1e11  # Initial guess for f_a in GeV
        fa_solution2, = fsolve(calc_f_a, fa_initial_guess, args=(m_a, n, 1, 1.1))
        if n == 0:
            if fa_solution2/factor > f_a_limit:
                break
            f_a_values_approx[n].append(fa_solution2/factor)
        else:
            if fa_solution2/factor > f_a_limit:
                break
            f_a_values_approx[n].append(fa_solution2/factor)


# Plotting
plt.figure(figsize=(8, 6))

# Plot f_a vs ma for different n values
for i, n in enumerate(n_values):
    colors = ['blue','green', 'purple'] 
    plt.loglog(m_a_values[:len(f_a_values[n])], f_a_values[n], c = colors[i] , label=f'n = {n}')
    plt.loglog(m_a_values[:len(f_a_values_approx[n])], f_a_values_approx[n], "--",c = colors[i], label=f'n = {n} (approx)')

# Adding the theoretical curve
mid_index = len(m_a_values) // 2  # Integer division for proper slicing
m_a_half = m_a_values[:mid_index]  # Slice array correctly
f_a_QCD = 1e12 * (6e-6 / m_a_half)
plt.loglog(m_a_half, f_a_QCD, "--", c ="black", label="QCD_axion")
plt.axhline(y=f_a_limit, color='black') #, label=r'$f_a = 8.2 \times 10^{12}$')

plt.xlabel(r'$m_a$ (eV)')
plt.xlim(1e-16, 1e-2)
plt.ylim(1e9, 1e14)
plt.ylabel(r'$f_a$ (GeV)')
plt.title(r'Plot of $f_a$ vs $m_a$ for different $n$ values')
plt.legend()
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()
```
